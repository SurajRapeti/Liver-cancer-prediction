# LinearRegression

# LogisticRegression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model_lg = LogisticRegression()
```

```python
model_lg.fit(X_train, y_train)
```

```python
predicted_lg=model_lg.predict(X_test)
```

```python
model_lg.score(X_test,y_test)
```

```python
confusion_matrix(y_test, model_lg.predict(X_test))
```

```python
print(f"Accuracy is {round(accuracy_score(y_test, model_lg.predict(X_test))*100,2)}")
```

```python
import pickle
pickle.dump(model_lg, open('liver.pkl', 'wb'))
```

```python
comparison=pd.DataFrame({'Actual':y_test,'Predicted':predicted_lg})
comparison[0:10]
```
